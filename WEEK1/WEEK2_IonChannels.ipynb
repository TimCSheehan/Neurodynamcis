```python
# t_step = .1
# t = np.arange(0,16,t_step)
def n_t_markov(N):
#N = 1000 #- # of gates
    gates = np.zeros(N)
    n_open = np.zeros(len(t))

    for t_ind in range(len(t)):
        for gate_i in range(N):
            prob_sample = random.uniform(0,1)
            if gates[gate_i] == 0:
                gates[gate_i] = prob_sample < (t_step*An_f(Vm[t_ind]))
            else:
                gates[gate_i] = prob_sample < (1-t_step*Bn_f(Vm[t_ind]))
        n_open[t_ind] = np.mean(gates)*100
    return n_open

```

```python
###########################################################
## Problem 2: Morris-Lecar (barnacle muscle fiber) Model ##
###########################################################
# Constants
C_m  =   1.0 # membrane capacitance, in uF/cm^2
g_Ca =   1.1 # maximum conducances, in mS/cm^2
g_K  =   36
g_L  =   0.3
g_Na =   120
E_Ca = 100.0 # Nernst reversal potentials, in mV
E_K  = -12.0
E_L  = 10.613
E_Na = 115
# Channel gating kinetics
# Functions of membrane voltage
def m_infty(V): return (1.0 + sp.tanh((V + 1.0) / 15.0)) / 2.0
def w_infty(V): return (1.0 + sp.tanh(V / 30.0)) / 2.0
def tau_w(V):   return 5.0 / sp.cosh(V / 60.0)  # in ms

# Membrane currents (in uA/cm^2)
def I_Ca(V):    return g_Ca * m_infty(V) * (V - E_Ca)
def I_K(V, w):  return g_K  * w          * (V - E_K)
def I_L(V):     return g_L               * (V - E_L)

# External current
# step up 10 uA/cm^2 every 100ms
def I_ext(t): return 10*sp.floor(t/100)+10

# The time to integrate over and the overall current trace
t = sp.arange(0.0, 400.0, 0.1)
I = I_ext(t)
```

```python
def a_m(V): return (25-V)/(10*(np.exp((25-V)/10)-1))
def b_m(V): return 4*np.exp(-V/18)
def a_h(V): return 0.07*np.exp(-V/20)
def b_h(V): return 1/(np.exp((30-V)/10)+1)
def a_n(V): return (10-V)/(100*(np.exp((10-V)/10)-1))
def b_n(V): return 0.125*np.exp(-V/80)
```

```python
###########################################################################
## Problem 4: Morris-Lecar model with expanded channel gating dynamics   ##
###########################################################################
# Constants
C_m  = 1.0      # membrane capacitance, in uF/cm^2
g_Ca = 1.1      # maximum conducances, in mS/cm^2
g_K  = 2.0 
g_L  = 0.5 
E_Ca = 100      # Nernst reversal potentials, in mV
E_K  = -70 
E_L  = -50 

# Channel gating kinetics
# Functions of membrane voltage
def m_infty(V): return (1 + tanh((V + 1) / 15)) / 2
def w_infty(V): return (1 + tanh(V / 30)) / 2
def tau_w(V): return 5 / cosh(V / 60)              # in ms

# Membrane currents (in uA/cm^2)
def I_Ca(V): return g_Ca * m_infty(V) * (V - E_Ca)
def I_K(V, w): return g_K * w * (V - E_K)
def I_L(V): return g_L * (V - E_L)
def I_ext(t): return 40+floor(t/20)

t_start = 0
t_stop = 200
t_step = 0.1
t = sp.arange(t_start, t_stop, t_step)
```

```python
##################################%
# 4.1 Linear solution (ODE solver)
##################################%
# Differential equation for X(1)=Vm and X(2)=w
def dXdt(X,t):
    V, w = X
    dVdt = (I_ext(t) - I_Ca(V) - I_K(V, w) - I_L(V)) / C_m
    dwdt = (w_infty(V) - w) / tau_w(V)
    return dVdt, dwdt

X = odeint(dXdt, [-70, 0], t)

Vm_linear = X[:,0] # the first column is the V values
w_linear  = X[:,1] # the second column is the w values

plot(t, Vm_linear)
plot(t, w_linear)
plot(t, I_ext(t))
```

```python
###################################
## 4.2 Stochastic solution (Markov)
###################################
def alpha_w(Vm): return w_infty(Vm) / tau_w(Vm)
def beta_w(Vm): return (1/tau_w(Vm)) - alpha_w(Vm)

N = 5000                      # number of gates
gates = zeros((N))            # all gates start in state 0
Vm_stoch = zeros(size(t))     # store Vm
w_stoch = zeros(size(t))      # store the fraction of gates open

Vm_current = -70
w_current = 0

for t_index in range(0,len(t)):

    # ODE solver
    def dVdt(Vm, t):
        return (I_ext(t) - I_Ca(Vm) - I_K(Vm, w_current) - I_L(Vm)) / C_m
    Vm = odeint(dVdt, Vm_current, sp.arange(t[t_index],t[t_index]+t_step,t_step/10))

    Vm_current = Vm[-1]
    Vm_stoch[t_index] = Vm_current
    
    for gate_index in range(0,len(gates)):
        prob_sample = rand()        # probabilistic sample
        if gates[gate_index]==0:
            # transition from state 0->1
            gates[gate_index] = prob_sample < t_step*alpha_w(Vm_current);
        else:
            # remain in state 1
            gates[gate_index] = prob_sample < (1-t_step*beta_w(Vm_current));

    w_current = sum(gates)/N
    w_stoch[t_index] = w_current

plot(t, Vm_stoch)
show()
plot(t, w_stoch)
```
